```python
import pandas as p
import os
import numpy as n
import matplotlib.pyplot as pl
```

```python
base_dir = os.getcwd()  # current working directory
data_path = os.path.abspath(os.path.join(base_dir, "..", "data", "cleaned_dataset.xlsx"))
dataset = p.read_excel(data_path)
dataset.head(5)
```

```python
dataset['InvoiceDate'] = p.to_datetime(dataset['InvoiceDate'])


now = dataset['InvoiceDate'].max() + p.Timedelta(days=1)
rfm = dataset.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda x: (now - x.max()).days),
    Frequency=('InvoiceDate', 'count'),
    Monetary=('Total_amount', 'sum')
).reset_index()

```

```python
base_dir = os.getcwd()  # current folder (e.g., notebooks)
save_path = os.path.abspath(os.path.join(base_dir, "..", "data", "RFM.xlsx"))
rfm.to_excel(save_path, index=False)
```
